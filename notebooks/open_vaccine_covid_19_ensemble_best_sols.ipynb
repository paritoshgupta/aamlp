```python
import os
import pandas as pd
os.chdir("/Users/paritoshgupta/Downloads/")
```

```python
df1 = pd.read_csv("submission (1).csv")
df2 = pd.read_csv("submission (2).csv")
```

```python
df1.head()
raw_cols = df1.columns
```

```python
df2.head()
```

```python
df1.columns = ['id_seqpos', 'reactivity_1', 'deg_Mg_pH10_1', 'deg_pH10_1', 'deg_Mg_50C_1',
       'deg_50C_1']
df2.columns = ['id_seqpos', 'reactivity_2', 'deg_Mg_pH10_2', 'deg_pH10_2', 'deg_Mg_50C_2',
       'deg_50C_2']
```

```python
out = pd.merge(df1, df2, how="inner", on ="id_seqpos")
```

```python
out
```

```python
out['reactivity'] = 0.6 * out['reactivity_1'] + 0.4 * out['reactivity_2']
out['deg_Mg_pH10'] = 0.65 * out['deg_Mg_pH10_1'] + 0.35 * out['deg_Mg_pH10_2']
out['deg_pH10'] = 0.7 * out['deg_pH10_1'] + 0.3 * out['deg_pH10_2']
out['deg_Mg_50C'] = 0.75 * out['deg_Mg_50C_1'] + 0.25 * out['deg_Mg_50C_2']
out['deg_50C'] = 0.7 * out['deg_50C_1'] + 0.3 * out['deg_50C_2']
```

```python
out[raw_cols].to_csv("submission_new.csv", index=False)
```
